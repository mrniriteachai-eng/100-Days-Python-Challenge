## ATM MINI PROJECT

```python
class ATM:

    def __init__(self):
        self.__balance = 5000
        self.pin = "1234"

    def check_pin(self):
        user_pin = input("Enter pin: ")

        if user_pin == self.pin:
            print("Pin correct")
            return True

        else:
            print("Wrong pin")
            return False

    def deposit(self, amount):
        self.__balance += amount
        print(f"Deposited Rs.{amount}")

    def withdraw(self, amount):

        if amount > self.__balance:
            print("Insufficient balance")

        else:
            self.__balance -= amount
            print(f"Withdraw Rs.{amount}")

    def show_balance(self):
        print(f"Current Balance: Rs.{self.__balance}")


atm = ATM()

if atm.check_pin():

    while True:

        print("\n1. Deposit")
        print("2. Withdraw")
        print("3. Balance Check")
        print("4. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            amount = int(input("Enter amount: "))
            atm.deposit(amount)

        elif choice == "2":
            amount = int(input("Enter amount: "))
            atm.withdraw(amount)

        elif choice == "3":
            atm.show_balance()

        elif choice == "4":
            print("Thank you")
            break
    

        else:
            print("Invalid choice")
```
