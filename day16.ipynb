smart id validation system

```python
users = []
n = int(input("how many user?"))

for i in range(n):

    name = input("enter name:")
    age = int(input("enter a age:"))
    email = input("enter wmail:")

    if len(name) < 3:
        print("invalid name")

    elif age < 18:
        print("underage")

    elif "@" not in email:
        print("invaalid email")

        users.append(name)
        print("user added")

        for user in users:
            print(user)



```
