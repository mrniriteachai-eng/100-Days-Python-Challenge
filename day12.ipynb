## DICTIONARY

```python
student = {
    "name": "ram",
    "age" : 18,
    "marks" : 90
}
print(student)
```

## ACCESS DATA

```python
print(student["name"])
print(student["marks"])
```

## ADD

```python
student["city"] = "kathmandu"
student ["marks"] = 95
```

## DELETE 

```python
del student["age"]
```

## LOOP

```python
for key in student:
    print(key, student[key])
```
