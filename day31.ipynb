## ENCAPSULATION IN PYTHON

## first simple code

```python
class Student:

    def __init__(self):
        self.name = "nirmal"
        self.age = 18


obj = Student()

print(obj.name)
print(obj.age)
```

## public variable

```python
class Bank:

    def __init__(self):
        self.balance = 500000


obj = Bank()

print(obj.balance)
```

## protected variable

```python
class Bank:

    def __init__(self):
        self._balance = 500000


obj = Bank()

print(obj._balance)
```

## private variable

```python
class Bank:

    def __init__(self):
        self.__balance = 500000


obj = Bank()

print(obj._balance)
```
