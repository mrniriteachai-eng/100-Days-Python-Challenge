## DIGITAL MEMORY SYSTEM

```python
def add_memory():
    data = input("Enter your memory: ")
    with open("vault.txt", "a") as file:
        file.write(data + "\n")
    print("Memory saved")


def view_memory():
    try:
        with open("vault.txt", "r") as file:
            print("\nYour memories:")
            print(file.read())
    except FileNotFoundError:
        print("No memories found")


def delete_memory():
    open("vault.txt", "w").close()
    print("All memories deleted")


while True:
    print("\n--- DIGITAL VAULT ---")
    print("1. Add memory")
    print("2. View memory")
    print("3. Delete memory")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_memory()

    elif choice == "2":
        view_memory()

    elif choice == "3":
        delete_memory()

    elif choice == "4":
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```
