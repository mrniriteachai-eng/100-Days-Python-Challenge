## basic code

```python
class Father:

    def house(self):
        print("father has a big house")

    def car(self):
        print("father has a car")


class Son(Father):
    pass


obj = Son()

obj.house()
obj.car()
```

## vehicle example

```python
class Vehicle:

    def start(self):
        print("hjjsj")


class Bike(Vehicle):
    pass


class Car(Vehicle):
    pass


b1 = Bike()
c1 = Car()

b1.start()
c1.start()
```
