```python
students = {}

def add_students():
    name = input("Enter student name: ")
    marks = int(input("Enter marks: "))

    students[name] = marks
    print("Student added successfully")


def show_students():
    if not students:
        print("No student found")
    else:
        for name, marks in students.items():
            print(name, ":", marks)


def calculate_average():
    if not students:
        print("No data")
    else:
        total = sum(students.values())
        avg = total / len(students)
        print("Average marks:", avg)


def find_topper():
    if not students:
        print("No data")
    else:
        topper = max(students, key=students.get)
        print("Topper:", topper, "with", students[topper])


def search_student():
    name = input("Enter name to search: ")

    if name in students:
        print(name, "marks:", students[name])
    else:
        print("Student not found")


while True:
    print("\n1. Add Student")
    print("2. Show Students")
    print("3. Average")
    print("4. Topper")
    print("5. Search")
    print("6. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        add_students()

    elif choice == "2":
        show_students()

    elif choice == "3":
        calculate_average()

    elif choice == "4":
        find_topper()

    elif choice == "5":
        search_student()

    elif choice == "6":
        print("Exiting...")
        break

    else:
        print("Invalid choice")
```
