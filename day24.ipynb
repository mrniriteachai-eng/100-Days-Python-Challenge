###SMART STUDY TIMER SYSTEM

```python
import time

study = int(input("enter study time:"))
break_time = int(input("enter break time"))


print("focus mode started")

while study:
    mins, secs = divmod(study, 60)
    print(f"{mins}: {secs}")
    time.sleep(1)
    study -=1

    print("study session compele")

    print("break time start")

    
while break_time:
    mins, secs = divmod(break_time, 60)
    print(f"{mins}: {secs}")
    time.sleep(1)
    break_time -= 1

    print("back to work stay consistent")
```
