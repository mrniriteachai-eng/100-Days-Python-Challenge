## CONSTRUCTOR  -- INIT--

```python
class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age


student1 = Student("nirmal", 18)

# print
print(student1.name)

print(student1.age)
```

## gaming character

```python
class Player:

    def __init__(self, name, health):
        self.name = name
        self.health = health



player1 = Player("nirmal", 18)


print(player1.name)

print(player1.health)
```
