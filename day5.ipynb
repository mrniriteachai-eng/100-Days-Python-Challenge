## IF ELSE
- if = yadi
- else = nabhay

## BASIC SYNTAX

```python
age = 18
if age >= 18:
    print("you can vote")
else:
    print("you cannot vote")
 
```

## REAL EXAMPLE

```python
num = int(input("enter a number"))

if num > 0:
    print("positive number")
else:
    print("negative number")
```

## ELIF

```python
marks = int(input("enter your marks:"))

if marks >= 80:
    print("grade A")
elif marks >= 60:
    print("grade B")
else:
    print("fail")
   
```
