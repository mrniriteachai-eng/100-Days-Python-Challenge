## CANNOT MODIFY

```python
t = [10, 20, 30]
t[0] = 100

print(t)
```

```python
t = [100, 20, 30]

for i in t:
    print(i)
```

## LIST SLICING

```python
num = [1, 2, 3, 4, 5]
print(num[1:4])
```

## LIST COMPREHENSION 

```python
nums = [1, 2, 3, 4]

squares = [x*x for x in nums]
print(squares)
```
