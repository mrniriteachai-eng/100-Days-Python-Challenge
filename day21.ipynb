NUMBER GUESSING GAME

```python
import random
number = random.randint(1,100)
guess = 0
attempts = 0
print("welcome to number guessing game")
print("guess a number between 1 and 100")
while guess != number:
    guess= int(input("enter your guess"))
    attempts +=1

    if guess < number:
        print("too low")

    elif guess > number:
        print("too high")
    else:
        print("correct")
        print("attempts:",attempts)
```
