```python
class students:
    name = "nirmal"
    age = 18

s1 = students()

print(s1.name)
print(s1.age)
```

```python
class students:
    name = "unknown"
    age = 0

    s1 = students ()
    s2 = students()

    s1.name = "nirmal"
    s1.age = 18

    s2.name = "saud"
    s2.age = "19"

    print(s1.name, s1.age)
    
    print(s2.name, s2.age)
```
