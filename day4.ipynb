## TYPE OF OPERATOR
1) Arithmetic operator



```python
a = 10
b = 5
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

2)  COMPARISION OPERATOR
-  compair garne

```python
a = 10
b = 5
print(a > b)
print(a < b)
print(a == b)
```

3) LOGICAL OPERATOR
- NOT OR AND 

```python
a = 10
print(a > 5 and a < 20)
print(a > 5 or a < 5)
print(not(a > 5  ))


```
